To remove outliers from 'sepal width (cm)'

```python
q1 = df['SepalWidthCm'].quantile(0.25)
q3 = df['SepalWidthCm'].quantile(0.75)
iqr = q3 - q1
df = df[(df['SepalWidthCm'] >= q1 - 1.5 * iqr) & (df['SepalWidthCm'] <= q3 + 1.5 * iqr)]
df.shape # To find out the number of rows and columns after outlier treatment
```

Blocplot for sepal width (cm) after outlier treatment

```python
sns.boxplot(y=df['SepalWidthCm'])
plt.show()
```

Splitting the data into train and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Species",axis=1)
y = df["Species"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state= 1)
```

Defining an object for DTC and fitting for whole dataset

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=3, min_samples_leaf=10, random_state=1 )
dt.fit(X, y)
```

Creating an object for DTC and applying it to the training dataset

```python
dt = DecisionTreeClassifier(random_state=1)
dt.fit(X_train, y_train)

y_pred_train = dt.predict(X_train)
y_pred = dt.predict(X_test)
y_prob = dt.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy of Decision Tree-Train: ', accuracy_score(y_pred_train, y_train))
print('Accuracy of Decision Tree-Test: ', accuracy_score(y_pred, y_test))
```

The output values suggest that the Decision Tree model performed perfectly on both the training dataset and the test dataset. An accuracy score of 1.0 indicates 100% accuracy, meaning that the model correctly predicted all the labels in both the training and test datasets. This could be a sign that the model may be overfitting the training data, achieving high accuracy on data it has seen before but potentially struggling with new, unseen data. It's essential to evaluate the model's performance on new data or consider other metrics to get a comprehensive understanding of its effectiveness.

Adjusting the settings of a Decision Tree Classifier to make it work better

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=1)

params = {'max_depth' : [2,3,4,5],
        'min_samples_split': [2,3,4,5],
        'min_samples_leaf': [1,2,3,4,5]}

gsearch = GridSearchCV(dt, param_grid=params, cv=3)

gsearch.fit(X,y)

gsearch.best_params_
```

```python
# Passing best parameter for the Hyperparameter Tuning
dt = DecisionTreeClassifier(**gsearch.best_params_, random_state=1)

dt.fit(X_train, y_train)

y_pred_train = dt.predict(X_train)
y_prob_train = dt.predict_proba(X_train)[:,1]

y_pred = dt.predict(X_test)
y_prob = dt.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print('Confusion Matrix - Train:','\n',confusion_matrix(y_train,y_pred_train))
print('\n','Confusion Matrix - Test:','\n',confusion_matrix(y_test,y_pred))
```

Classification for test after hyperparameter tuning

```python
print(classification_report(y_test,y_pred))
```

```python
print('Accuracy of Decision Tree-Train: ', accuracy_score(y_pred_train, y_train))
print('Accuracy of Decision Tree-Test: ', accuracy_score(y_pred, y_test))
```

The output indicates that the Decision Tree classifier has achieved perfect accuracy of 100% on both the training and test datasets. An accuracy score of 1.0 suggests that the model correctly classified all instances in both datasets. While high accuracy is desirable, it's essential to interpret these results with caution. A perfect accuracy score on the training set may indicate overfitting, where the model has memorized the training data but may not generalize well to unseen data. Therefore, it is advisable to further evaluate the model's performance using other metrics and consider fine-tuning its parameters to achieve a balance between training performance and generalization to new data.
